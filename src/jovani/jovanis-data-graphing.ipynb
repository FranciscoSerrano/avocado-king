```python
# Import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# File to Load (Remember to change the path if needed.)
avocado_data_to_load = os.path.join("./data/interim", "avocado.pkl")

# Load to data frame from a pickle file.
vis_guacate_df = pd.read_pickle(avocado_data_to_load)
vis_guacate_df
```

```python
# Testing out some methods to display matplotlib visuals
avo_type = vis_guacate_df.groupby("type")["Total Volume"].agg("sum")
print(avo_type["conventional"])
print(avo_type["organic"])
total = avo_type["conventional"] + avo_type["organic"]
print(avo_type["conventional"]/total)
print(avo_type["organic"]/total)
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise of 'conventional' and 'organic':
# https://matplotlib.org/3.1.1/gallery/pie_and_polar_charts/pie_features.html
labels = "conventional", "organic"
sizes = [(avo_type["conventional"]/total), (avo_type["organic"]/total)]
explode = (0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax.axis("equal")
```
